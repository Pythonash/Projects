송예나 교수님이 데이터 정보좀 달라함

상관계수랑 p-value, 기초통계량 제공


결측값 있는 데이터셋(1)과 결측값 없고 표준화된 데이터셋(2) 둘다 드리기로함.

```python
import os

os.chdir('/content/drive/MyDrive/Green Index Paper')
```

```python
import pandas as pd

import scipy.stats as st

import numpy as np
```

```python
df1=pd.read_csv('./부산그린인덱스_방법두가지_결측값있음.csv')
df2=pd.read_csv('./부산_결측값없고_표준화.csv')
```

```python
df1
```

```python
df2
```

```python
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
df1.describe().T
```

```python
df2.describe().T
```

```python
st.pearsonr(df1['HGVI_2'], df1['Price'])
```

```python
st.pearsonr(df1['Price'], df1['HGVI_2'])
```

```python
df1.corr()
```

```python
df2.corr()
```

```python
co_1 = df1.drop(['HGVI_1'],axis=1).corr()
co_2 = df2.corr()

pv_1 = df1.drop(['HGVI_1'],axis=1).corr()
pv_2 = df2.corr()

co_1[co_1.columns] = np.NaN
co_2[co_2.columns] = np.NaN

pv_1[pv_1.columns] = np.NaN
pv_2[pv_2.columns] = np.NaN

co_1 = co_1.reset_index()
co_2 = co_2.reset_index()

pv_1 = pv_1.reset_index()
pv_2 = pv_2.reset_index()
```

```python
co_1
```

```python
for a in df1.drop(['HGVI_1'],axis=1):
  corr_1 = []
  pvalue_1 = []
  for b in df1.drop(['HGVI_1'],axis=1):
    co, pv = st.pearsonr(df1[a], df1[b])
    corr_1.append(round(co,4))
    pvalue_1.append(round(pv,4))
  co_1[a] = corr_1
  pv_1[a] = pvalue_1
  print(co_1)
```

```python
for a in df2:
  corr_2 = []
  pvalue_2 = []
  for b in df2:
    co, pv = st.pearsonr(df2[a], df2[b])
    corr_2.append(round(co,4))
    pvalue_2.append(round(pv,4))
  co_2[a] = corr_2
  pv_2[a] = pvalue_2
  print(co_2)
```

```python
co_1
```

```python
co_2
```

결측값 있는 데이터셋 상관계수와 유의수준의 경우 HGVI_1이 빠짐

```python
co_1.to_csv('./결측값있는데이터셋상관계수.csv',encoding='utf-8-sig', index=False)
co_2.to_csv('./결측값없는데이터셋상관계수.csv',encoding='utf-8-sig', index=False)

pv_1.to_csv('./결측값있는데이터셋유의수준.csv',encoding='utf-8-sig', index=False)
pv_2.to_csv('./결측값없는데이터셋유의수준.csv',encoding='utf-8-sig', index=False)
```

```python
ds1 = df1.describe().T

ds1[ds1.columns] = round(ds1, 3)

ds1[['count','mean','std','min','max']].to_csv('./결측값있는데이터셋기초통계량.csv',encoding='utf-8-sig')

```

## HGVI 결측값 채워서 한번 해보기


첫 번째 방법으로 수집한 HGVI의 경우 결측값이 존재하기 때문에 이를 선형회귀, KNN으로 채워서 한번 해보려 한다.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

df = pd.read_csv('/content/drive/MyDrive/Green Index Paper/부산_결측값있음_방법두가지.csv')
df
```

```python
var = df.drop(['x','y','Price','HGVI_2'],axis=1)

var = var.dropna().reset_index().drop(['index'],axis=1)

var
```

선형회귀 피팅

```python
model = LinearRegression()

model.fit(var.drop(['HGVI_1'],axis=1), var['HGVI_1'])
```

최근접이웃 피팅

```python
model2=KNeighborsRegressor()
model2.fit(var.drop(['HGVI_1'],axis=1), var['HGVI_1'])
```

피팅시킬 결측값 모아주기

```python
df_null = df[df['HGVI_1'].isnull() != False].drop(['Price','x','y','HGVI_1','HGVI_2'],axis=1)
df_null
```

```python
print(df_null.columns)
print(var.columns)
```

```python
linear = pd.DataFrame(model.predict(df_null))
linear.index = df_null.index
linear.columns = ['HGVI_1']
linear
```

```python
knn = pd.DataFrame(model2.predict(df_null))
knn.index = df_null.index
knn.columns = ['HGVI_1']
knn
```

```python
linear.describe()
```

```python
knn.describe()
```

```python
df['HGVI_1_linear'] = df['HGVI_1']
df['HGVI_1_knn'] = df['HGVI_1']

df
```

```python
df['HGVI_1_linear'].loc[df_null.index] = linear['HGVI_1']
df['HGVI_1_knn'].loc[df_null.index] = knn['HGVI_1']

df
```

```python
df.isnull().sum()
```

```python
df.to_csv('/content/drive/MyDrive/Green Index Paper/LR_KNN_결측값처리.csv', index=False, encoding='utf-8-sig')
```
