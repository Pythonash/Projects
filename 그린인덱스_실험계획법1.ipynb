```python
co_1 = df1.drop(['HGVI_1'],axis=1).corr()
co_2 = df2.corr()

pv_1 = df1.drop(['HGVI_1'],axis=1).corr()
pv_2 = df2.corr()

co_1[co_1.columns] = np.NaN
co_2[co_2.columns] = np.NaN

pv_1[pv_1.columns] = np.NaN
pv_2[pv_2.columns] = np.NaN

co_1 = co_1.reset_index()
co_2 = co_2.reset_index()

pv_1 = pv_1.reset_index()
pv_2 = pv_2.reset_index()
```

```python
co_1
```

```python
for a in df1.drop(['HGVI_1'],axis=1):
  corr_1 = []
  pvalue_1 = []
  for b in df1.drop(['HGVI_1'],axis=1):
    co, pv = st.pearsonr(df1[a], df1[b])
    corr_1.append(round(co,4))
    pvalue_1.append(round(pv,4))
  co_1[a] = corr_1
  pv_1[a] = pvalue_1
  print(co_1)
```

```python
for a in df2:
  corr_2 = []
  pvalue_2 = []
  for b in df2:
    co, pv = st.pearsonr(df2[a], df2[b])
    corr_2.append(round(co,4))
    pvalue_2.append(round(pv,4))
  co_2[a] = corr_2
  pv_2[a] = pvalue_2
  print(co_2)
```

- Dist, green, and water는 (-)로 나와야 하고, 환경 변수가 많이 들어가야 함

- HGVI는 (+)로 나오면 좋은데 이게 green 하고 대체인지 보충관계인지 알려면 빼고 넣어보고 같이 넣어본 다음에 모델 파워가 쎄지는 지(만약 그렇다면 보충관계) 등등 알아봐야함

- 방법 2의 경우 거리 기준은 50개가 적당한 것 같음

- 아파트를 기준으로 도로의 길이 합(이건 되면 하는건데, 도로 밀도 데이터 구하기, 동별 도로 밀도, 도로 길이, 도로 연장 등의 데이터)

- 관련한 가설에 대해서도 생각해보기

- 아파트 데이터셋 보내드리기

- 이 변수가 무엇인지 생각하면서 돌리기

- 대전을 기준으로 한번 해보기

- 대전에 고학력자가 모여사는 동네가 있을 수 있다

- 그래서 대전을 가지고 한번 해보는데, 이에 대한 여러 가지 논문(학벌에 대한?) 읽어 보고 어떤 변수가 또 추가되면 좋을지 혹은 데이터라도 찾아서 넣어볼 수 있을지 등 해보기

그 전에, 데이터를 병합해 주는 작업이 필요하다 ( 인구나 성별과 같은 자료를 합쳐주어야 해서)

```python
import pandas as pd

df1 = pd.read_csv('/content/drive/MyDrive/Green Index Paper/busan_interpolation_0228.csv')
df2 = pd.read_excel('/content/drive/MyDrive/다운로드/0304_Busan_imputed_road_length.xlsx')
```

```python
df1
```

```python
import numpy as np
df1['Shape_area'] = np.NaN
df1['Road_length'] = np.NaN
df1['Road_density'] = np.NaN
num = 1 
for xx, yy, ind in zip(df1['x'],df1['y'], df1.index):
  xx = round(xx,6)
  yy = round(yy,6)
  for xxx, yyy, shape, length, density in zip(df2['x'], df2['y'], df2['SHAPE_AREA'], df2['Road_length'], df2['Road_density']):
    xxx =round(xxx,6)
    yyy = round(yyy,6)
    if (xxx == xx) & (yyy == yy):
      print("위치좌표 일치!!, 현재 인덱스 {}".format(num))
      df1['Shape_area'].iloc[ind] = shape
      df1['Road_length'].iloc[ind] = length
      df1['Road_density'].iloc[ind] = density
  num+=1
```
